**Merging dataframes**

Merging Inquiry and Users

```python
df = pd.merge(inquiry,users,on=["user_id","created_at"],how="left") 
```

```python
df = df[['inquiry_id', 'user_id', 'procedure_id', 'created_at', 'clinic_id','country']]
```

Merging with Procedures

```python
df = pd.merge(df,procedures,on="procedure_id",how="left")
df.drop(['created_at_y','updated_at'],axis=1,inplace=True)
```

Merging with Clinics

```python
df = pd.merge(df,clinics[["clinic_id","country"]],on="clinic_id",how="left")
```

```python
df.rename(columns={"created_at_x":"inquiry_date","country_x":"user_country","country_y":"clinic_country"},inplace=True)
```

Merging with Bookings

```python
df = pd.merge(df,bookings,on="inquiry_id",how="left")
# some inquiry_ids match with many booking_ids
```

```python
df.rename(columns={"created_at":"booking_date"},inplace=True)
```

```python
df.isnull().sum()
```

```python
booking = df[df["booking_id"].notnull()]
```

```python
df.fillna("no booking",inplace=True)
```

```python
df.procedure_name.value_counts()
```

**Adding Values and Cost per Inquirer**

```python
value={"Hip Replacement":10000,"Veneers":300,"Teeth Whitening":250,"Breast Implants":5000,"Cancer Treatment":9000,"Laser Eye Surgery":1000}
cpi = {"Hip Replacement":250,"Veneers":45,"Teeth Whitening":50,"Breast Implants":100,"Cancer Treatment":300,"Laser Eye Surgery":75}
```

```python
df["value"] = df["procedure_name"].map(value)
df["cpi"] = df["procedure_name"].map(cpi)
```

```python
df.info()
```

**Conversion Rate Among Inquiries and Bookings**

```python
conversion_all = pd.DataFrame((df.groupby(["status"])["inquiry_id"].count() / len(df)) * 100).reset_index()
```

```python
conversion_booking = pd.DataFrame((df[df["status"]!="no booking"].groupby(["status"])["booking_id"].count() / len(df[df["status"]!="no booking"])) * 100).reset_index()
```

```python
sns.set_theme(style="darkgrid")
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Conversion Rate Among Inquiries and Bookings')

sns.barplot(ax=axes[0],x="inquiry_id",y="status",data=conversion_all,palette=["C1", "C2", "C3", "C0"])
axes[0].set_title("Percentage of Completion in Inquiries")
axes[0].set(xlabel="Percentage")


sns.barplot(ax=axes[1],x="booking_id",y="status",data=conversion_booking,palette=["C1", "C2", "C0"])
axes[1].set_title("Percentage of Completion in Bookings")
axes[1].set(xlabel="Percentage")

plt.tight_layout()
```

Overall 22% of the inquiries and 43% of the bookings are being completed

**Conversion Rate By Country**

```python
conversion_country = (pd.crosstab(df["user_country"],df["status"]).apply(lambda x: x / pd.crosstab(df["user_country"],df["status"]).sum(axis=1)) * 100).stack().reset_index()
```

```python
conversion_country_booking = (pd.crosstab(df[df["status"]!="no booking"]["user_country"],df[df["status"]!="no booking"]["status"]).apply(lambda x: x / pd.crosstab(df[df["status"]!="no booking"]["user_country"],df[df["status"]!="no booking"]["status"]).sum(axis=1)) * 100).stack().reset_index()
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Conversion by Country')

sns.barplot(ax=axes[0],x="user_country",y=0,data=conversion_country,hue="status",palette=["C1", "C2", "C3", "C0"])
axes[0].set_title("Percentage of Completion in Inquiries for Countries")
axes[0].set(xlabel="Countries",ylabel="Percentage")



sns.barplot(ax=axes[1],x="user_country",y=0,data=conversion_country_booking,hue="status",palette=["C1", "C2", "C0"])
axes[1].set_title("Percentage of Completion in Bookings for Countries")
axes[1].set(xlabel="Countries",ylabel="Percentage")

plt.tight_layout()
```

No country would be considered as outlier in terms of the conversion rate, yet the cancellation from United Kingdom is lower than other countries

**Inquiries and Bookings by Date**

Inquiries

```python
df["inquiry_month"] = df["inquiry_date"].dt.strftime("%b %Y")
inquiry_sort = df.groupby(["inquiry_month"])["status"].count().reset_index()
months_2014 = [x[:3] + " 2014" for x in calendar.month_name[1:]]
inquiry_sort['inquiry_month'] = pd.Categorical(inquiry_sort['inquiry_month'], categories=months_2014, ordered=True)
inquiry_sort.sort_values(by="inquiry_month",inplace=True)
```

Bookings

```python
booking["booking_month"] = booking["booking_date"].dt.strftime("%b %Y")
months_2015 = [x[:3] + " 2015" for x in calendar.month_name[1:]]
bookings_sort = booking.groupby(["booking_month","status"])["booking_id"].count().reset_index()
bookings_sort['booking_month'] = pd.Categorical(bookings_sort['booking_month'], categories=months_2014 + months_2015, ordered=True)
bookings_sort.sort_values(by="booking_month",inplace=True)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Inquiries and Bookings by Date')


sns.lineplot(ax=axes[0],x="inquiry_month",y="status",data=inquiry_sort)
plt.draw()
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=30)
axes[0].set(xlabel="Date",ylabel="Inquiry")
axes[0].set_title("Inquiries by Date")


sns.lineplot(ax=axes[1],x="booking_month",y="booking_id",data=bookings_sort,hue="status",palette=["C1", "C2", "C0"])
plt.draw()
axes[1].set_xticklabels(axes[1].get_xticklabels(),rotation=30)
axes[1].set(xlabel="Date",ylabel="Booking")
axes[1].set_title("Bookings by Date")


plt.show()
```

There is an increasing trend in the number of Inquiries between April and September.

In contrast to Inquiries, Bookings reached to peak in October-November. There is a certain amount of time between inquiry and booking, also observing bookings in 2015 indicates that

**Average Profit**

```python
booking["value"] = booking["procedure_name"].map(value) #add values to booking dataframe based on procedure_name
booking["cpi"] = booking["procedure_name"].map(cpi) #add cpi to booking dataframe based on procedure_name
booking["inquiry_month"] = booking["inquiry_date"].dt.strftime("%b %Y")
```

```python
cash = pd.concat([df.groupby(['inquiry_month'])['cpi'].sum() / df.groupby(['inquiry_month'])['inquiry_id'].nunique(),(booking[booking.status=='completed'].groupby(['booking_month'])['value'].sum() * 0.30) / booking[booking.status=='completed']['booking_id'].nunique(),(booking[booking.status=='completed'].groupby(['inquiry_month'])['value'].sum() * 0.30) / booking[booking.status=='completed']['booking_id'].nunique()],axis=1).reset_index()
cash.columns = ['date','CPI','booking','inquiry']
cash['date'] = pd.Categorical(cash['date'], categories=months_2014 + months_2015, ordered=True)
cash.sort_values(by="date",inplace=True)
```

```python
fig, axes = plt.subplots(1, 2, figsize=(15, 5))
fig.suptitle('Profit by Booking and Inquiry Date')


sns.lineplot(ax=axes[0],x="date",y=(cash['booking'][:8] - cash['CPI'][:8]),data=cash)
plt.draw()
axes[0].set_xticklabels(axes[0].get_xticklabels(),rotation=30)
axes[0].set(xlabel="Date",ylabel="Average Profit")
axes[0].set_title("Average Profit by Booking Date")


sns.lineplot(ax=axes[1],x="date",y=(cash['inquiry'][:8] - cash['CPI'][:8]),data=cash)
plt.draw()
axes[1].set_xticklabels(axes[1].get_xticklabels(),rotation=30)
axes[1].set(xlabel="Date",ylabel="Average Profit")
axes[1].set_title("Average Profit by Inquiry Date")

plt.show()
```

It is obvious that users who made inquiry complete the booking later. Between **April and October 2014** we can see that average CPI exceeds the average commission received. However, it should be also noted that there are **710** scheduled bookings and this is almost as high as completed bookings that are **762** right now.


When we look at the profits aggregated in **inquiry date** level it can be seen that in August and September 2014 commission received exceeded CPI. However as inquiries started to decrease after **September 2014** the profit by inquiry date also decreased along with it

**Logistic Regression for Booking Status**

```python
df_log = df[(df.status == 'completed') | (df.status == 'cancelled')][['booking_id','status','clinic_country','procedure_name']]
```

```python
df_log['status'] = np.where(df_log['status']=='completed',1,0)
```

```python
pd.crosstab(df_log['clinic_country'],df_log['status'])
```

```python
pd.crosstab(df_log['procedure_name'],df_log['status'])
```

```python
df_log[list(pd.get_dummies(df_log['clinic_country']).columns)] = pd.get_dummies(df_log['clinic_country'])
```

```python
df_log[list(pd.get_dummies(df_log['procedure_name']).columns)] = pd.get_dummies(df_log['procedure_name'])
```

```python
lst1 = list(pd.get_dummies(df_log['clinic_country']).columns)
lst1.remove('Malaysia')
lst2 = list(pd.get_dummies(df_log['procedure_name']).columns)
lst2.remove('Laser Eye Surgery')
```

```python
y = df_log['status']
X = pd.concat([df_log[lst1],df_log[lst2]],axis=1) 
X = sm.add_constant(X)
model1 = sm.Logit(y, X).fit()
model1.summary()
```

**Logistic Regression for Inquiry Status**

```python
df_log2 = df[['status','clinic_country','procedure_name']]
```

```python
df_log2['status'] = np.where(df_log2['status']=='no booking',0,1)
```

```python
pd.crosstab(df_log2['clinic_country'],df_log2['status'])
```

```python
pd.crosstab(df_log2['procedure_name'],df_log2['status'])
```

```python
df_log2[list(pd.get_dummies(df_log2['clinic_country']).columns)] = pd.get_dummies(df_log2['clinic_country'])
df_log2[list(pd.get_dummies(df_log2['procedure_name']).columns)] = pd.get_dummies(df_log2['procedure_name'])
```

```python
y = df_log2['status']
X = pd.concat([df_log2[lst1],df_log2[lst2]],axis=1) 
X = sm.add_constant(X)
model2 = sm.Logit(y, X).fit()
model2.summary()
```

**Comparison of Models**

```python
dfoutput = summary_col([model1,model2],stars=True,model_names=['Bookings','Inquiries'],info_dict={'N':lambda x: "{0:d}".format(int(x.nobs)),'R2':lambda x: "{:.2f}".format(x.prsquared)})
print(dfoutput)
```
